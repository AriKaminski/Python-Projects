```python
# make filename a global variable
# This file must already exist for this work!!!!!!
FILENAME = "contacts.csv"

def write_contacts(contacts_list):
    with open(FILENAME, "w") as file:
        for contact in contacts_list:
            file.write(contact + "\n")
            
def read_contacts():
    contacts_list = []
    with open(FILENAME, newline="") as file:
        for line in file:
            line = line.replace("\n", "")
            contacts_list.append(line)
    return(contacts_list)

def list_contacts(contacts_list):
    for i in range(len(contacts_list)):
        contact = contacts_list[i]
        print(str(i+1) + ". " + contact)       
    print()
    
def add(contacts_list):
    
    """The add funtion adds a contact to the contact list, contact is added at the 0 posistion"""
        
    name = input("Contact Name: ")
    email = input("Contact Email: ")
    number = input("Contact Phone Number: ")
    contact = []
    contact.append(name)
    contact.append(email)
    contact.append(number)
    contacts_list.append(contact)
    write_contacts(contacts_list)
    print(name + " was added to the contact list.\n")
    
def delete(contacts_list):
    
    """The delete function removes a contact from the list. 
    The posistion of the contact to be deleted is input by the user"""
    
    number = int(input("Contact number posistion: "))
    if number < 1 or number > len(contacts_list):
        print("Invalid contact number.\n")
    else:
        contacts = contacts_list.pop(number-1)
        write_contacts(contacts_list)
        print(contacts[0] + " was deleted from the contact list.\n")
        
def view(contacts_list):
    
    """The view function views all information about a given contact. 
    The posistion of the contact is input by the user"""
    
    number = int(input("Contact number posistion: "))
    if number < 1 or number > len(contacts_list):
        print("Invalid contact number.\n")
    else:
        contacty = contacts_list.pop(number - 1)
        print("Name )" + contacty[0])
        print("Email )" + contacty[1])
        print("Phone )" + contacty[2])
        print("\n")



def display_menu():
    print("The Contacts List Program\n")
    print("view - view a contact")
    print("list - List all contacts")
    print("add - Add a contact")
    print("del - Delete a contact")
    print("exit - Exit program")    
    
    

    
def main():
    
    """The main function is what is run"""
    
    contacts_list = read_contacts()
    display_menu()
    command = input("Enter input: ")
        
    while (command.lower() != "exit"):
        
        if command.lower() == "list":
            list_contacts(contacts_list)
            display_menu()
        elif command.lower() == "view":
            view(contacts_list)
            display_menu()
        elif command.lower() == "add":
            add(contacts_list)
            display_menu()
        elif command.lower() == "del":
            delete(contacts_list)
            display_menu()
        else:
            print("Not a valid input. \n")

        command = input("Enter input: \n" )
         
    print("Program finished")     
        
if __name__ == "__main__":
    main()
```
