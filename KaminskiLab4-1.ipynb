```python
#This is a program that displays the sum of even and odd integers
# Ari Kaminski

sum = 0
sum2 = 0
sum3 = 0

for i in range(0, 101, 2):
    sum = sum + i
    
print('The sum of all even integers between 1 and 100: ' + str(sum))
print()
    
for i in range(1, 101, 2):
    sum2 = sum2 + i

print('The sum of all odd integers between 1 and 100: ' + str(sum2))
print()

for i in range(1, 101):
    sum3 = sum3 + i
    
print('The sum of all odd integers between 1 and 100: ' + str(sum3))
```

```python
#This is a program that flips a coin 100 and displays results
#Ari Kaminski
import random

i=0

sum1 = 500
num_heads = 0
num_tails = 0


while i < 100:
    coin = random.choice(['heads', 'tails'])
    print("The coin landed on " + coin)
    i = i + 1

    if(coin == 'heads'):
        print('You WIN!')
        sum1 = sum1 + 10
        num_heads = num_heads + 1
    elif(coin == 'tails'):
        print('You LOSE!')
        sum1 = sum1 - 10
        num_tails = num_tails + 1
print()
print('Number of heads: ' + str(num_heads))
print('Number of tails: ' + str(num_tails))
print('You now have  $' + str(sum1) + ' in your wallet.')
```
