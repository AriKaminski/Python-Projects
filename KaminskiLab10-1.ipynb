```python
class Movie:
    

    #constructor that initilizes
    def __init__(self, name = "", year = 0):
        self.__name = name
        self.__year = year
    
    def setName(self):
        self.__name = name
    
    
    def setYear(self):
        self.__year = year
     
    
    def getName(self):
        return self.__name
    
    def getYear(self):
        return self.__year
    
    def getStr(self):
        print(self.__name + "  (" + str(self.__year) + ")")
        
        
    
def display_menu():
    print("COMMAND MENU")
    print("list - List all movies")
    print("add - Add a movie")
    print("del - Delete a movie")
    print("exit - exit program")
    
    
def list(movie_list):
    
    """The list function lists all movies"""
    
    print("MOVIES")
    i=0
    if len(movie_list) == 0:
        print("There are no contacts in the list.\n")
    else:
        for movie in movie_list:
            i += 1
            print(str(i) + ")"), movie.getStr()
    
def add(movie_list):
    
    """The add function adds a movie object"""
    
    movie_name = str(input("Enter movie name = "))
    movie_year = int(input("Enter movie year = "))
    new_movie = Movie(movie_name , movie_year)
    movie_list.append(new_movie)
    print(movie_name, " was added to the list")
    
def delete(movie_list):
    
    """The delete function del a movie object"""
    num = int(input("Enter number of movie to be deleted = "))
    num = (num - 1)
    movie_del = movie_list.pop(num)
    print(movie_del.getStr(), " was deleted from the list")
    
    
def main():
    
    movie_list =[Movie("Remember the Titans", 2000),
                Movie("An Old Man and A Gun", 2018),
                Movie("The Spy who Dumped Me", 2019)]
        
    display_menu()
    command = input("Enter Command: ")
        
    while (command.lower() != "exit"):
        
        if command.lower() == "list":
            list(movie_list)
            display_menu()
        elif command.lower() == "add":
            add(movie_list)
            display_menu()
        elif command.lower() == "del":
            delete(movie_list)
            display_menu()
        else:
            print("Not a valid command.  Please try again.\n")

        command = input("Enter Command: ")
         
    print("You'll be back!")
    
if __name__ == "__main__":
    main()
```
